```python
import pandas as pd
import numpy as np
```

```python
# load data
customer_churn=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
customer_churn.head(5)
```

```python
customer_churn.dtypes
```

```python
# check if there is null 
customer_churn.isnull().sum()
```

```python
# drop colimn customerID
customer_churn = customer_churn.drop(columns=['customerID'])
```

```python
customer_churn.head(5)
```

```python
# change object values to numeric
customer_churn['TotalCharges'] = pd.to_numeric(customer_churn['TotalCharges'], errors='coerce')
```

```python
customer_churn.head(5)
```

```python
# if there are NaN values changes them to median
customer_churn['TotalCharges'].fillna(customer_churn['TotalCharges'].median(), inplace=True)
```

```python
customer_churn.head(5)
```

```python
# change predicted value to numeric
customer_churn['Churn'] = customer_churn['Churn'].map({'No': 0, 'Yes': 1})
```

```python
# check if values are balanced
customer_churn['Churn'].value_counts(normalize=True)
```

```python
# assign X value, dropping the value we need to predict
X = customer_churn.drop('Churn', axis=1)
X = pd.get_dummies(X)  
y = customer_churn['Churn']
```

```python
# build SMOTE model to balance values
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
print("Before SMOTE:", X_train.shape, y_train.value_counts())
print("After SMOTE:", X_train_smote.shape, y_train_smote.value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=y_train_smote)
plt.title("Balance of classes after SMOTE")
plt.show()
```

```python
# build XGBoost model for prediction
from xgboost import XGBClassifier
model = XGBClassifier(random_state=42)
model.fit(X_train_smote, y_train_smote)
```

```python
y_pred = model.predict(X_test)
```

```python
# check the results
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
feature_importances = pd.Series(model.feature_importances_, index=X_train_smote.columns)
feature_importances.sort_values(ascending=False).head(10).plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.show()
```
